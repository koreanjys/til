## 여러 그래프 그리기

```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
```

```python
# 1번 방법
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
# 2번 방법

plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

```python
### 새로운 그래프 창을 생성함

x = np.arange(-4.5, 5, 0.5)
y1 = 2*x
y2 = x**2
y3 = -2*x
y4 = -x**2

plt.plot(x, y1)  # 그래프 생성 (default 값이 figure(1)인듯?)

plt.figure(1)  # 1번 그래프 창 생성

plt.plot(x, y2)  # 바로 위에 그래프 창에 그려진다.(1번)

plt.figure(2)  # 2번 그래프 창 생성

plt.plot(x, y3)  # 바로 위에 그래프 창에 그려진다.(2번)

plt.plot(x, y4)  # 마지막 figure(n) 에 그려진다.

plt.show()
```
