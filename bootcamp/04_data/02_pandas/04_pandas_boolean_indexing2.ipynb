```python
# 60세 넘는 사람들을 인덱싱  // .reset_index()
titan_df[titan_df['Age'] > 60]
titan_df[titan_df['Age'] > 60].reset_index()  # index값을 새로 추가
titan_df[titan_df['Age'] > 60].reset_index(drop=True)  # 기존 index 값은 없애며 새로운 index 추가
```

```python
# 성별이 여성인 승객의 '이름', '나이' 컬럼을 추출
titan_df[titan_df.Sex=='female'][['Name', 'Age']]
```

### 복합조건
- and 조건은 &
- or 조건은 |
- not 조건은 ~

### 연습문제

```python
# Q1 60세 초과(넘는다) 여성이고 선실등급 1등급에 자리했던 승객을 구해주세요... :: 2명

# 1번째 조건 변수를 준다
cond1 = titan_df.Age>60
cond2 = titan_df.Sex == 'female'
cond3 = titan_df.Pclass == 1
titan_df[cond1&cond2&cond3]

# 2번째 변수 없이 ()를 써서 한다. => () 없으면 여러 type이 섞여서 에러가 뜬다.
titan_df[(titan_df.Age > 60) & (titan_df.Sex == 'female') & (titan_df.Pclass == 1)]

# 3번째 []마다 조건을 넣어준다 (그런데 빨간 박스 에러 뜸)
titan_df[titan_df.Age > 60][titan_df.Sex == 'female'][titan_df.Pclass == 1]

# 4번째 eq, gt 등등 쓴다
titan_df[titan_df['Pclass'].eq(1) & titan_df.Sex.eq('female') & titan_df['Age'].gt(60)]
```

```python
# 위 문제를 부정문으로 => 이 방법은 사용하지 말기
titan_df[~cond1&~cond2&~cond3]
```
