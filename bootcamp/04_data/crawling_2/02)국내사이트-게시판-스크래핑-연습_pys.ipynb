```python
import requests as req
from bs4 import BeautifulSoup

url = 'https://tbever.com/%EA%B5%AD%EB%82%B4-%EC%9B%B9%EC%82%AC%EC%9D%B4%ED%8A%B8-%EB%9E%AD%ED%82%B9-top50/'

# HTML소스코드 가져오기.
html_web_rank = req.get(url).text 

# Parsing(파싱) - Beutifulsoup
soup_web_rank = BeautifulSoup(html_web_rank)
```

```python
# p 태그의 요소안에 span 하위의 a 태그의 요소를 찾는다.
rank_50 = soup_web_rank.select('p span a') #CSS를 활용해 찾는 방식
```

```python
# for문을 활용하여 전체 사이트를 리스트에 담기

result = []

for rank_i in rank_50:
    result.append(rank_i.get_text())
```

```python
# for 리스트 내포형식으로 담기

result1 = [rank_i.get_text() for rank_i in rank_50]
```

```python
# pandas df -> csv로 저장
import pandas as pd
result_df = pd.DataFrame(result1, columns=['web_site'])
result_df
```

```python
result_df.to_csv('./data_result/web_50.csv')
```
