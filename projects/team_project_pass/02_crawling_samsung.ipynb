```python
import pandas as pd
import os
from selenium import webdriver
```

```python
# 크롬으로 웹창 시작  # 한번에 4개 csv 파일 만들 수 있도록 정리 요망
browser = webdriver.Chrome()
# 증권 데이터 url 정의  (삼성=005930, SK하이닉스=000660, SK스퀘어=402340, DB하이텍=000990)
data_code = {'samsung': '005930', 'skhynix': '000660', 'sksquare': '402340', 'dbhitek': '000990'}
page = 1

for p in range(1, 37):
    url = f'https://finance.naver.com/item/sise_day.naver?code={data_code}&page={page}'
    # url 정보 가져오기
    browser.get(url)
    # 삼성 데이터 추출
    samsung_df = pd.read_html(browser.page_source)[0]
    # 모든 행의 요소가 결측치일 경우 행 삭제
    samsung_df.dropna(axis='index', how='all', inplace=True)
    # 전일비 culumn 삭제
    samsung_df.drop(['전일비'], axis=1, inplace=True)
    # csv파일로 저장
    data_info = 'samsung.csv'
    if os.path.exists(samsung_info):
        samsung_df.to_csv(samsung_info, encoding='utf-8-sig', index=False, mode='a', header=False)
    else:
        samsung_df.to_csv(samsung_info, encoding='utf-8-sig', index=False)
    print(f'{page}페이지 완료')
    page += 1
```
