#### PCA 했을 때

```python
from sklearn.model_selection import train_test_split

import pandas as pd

df = pd.read_csv("diabetes_data.csv", encoding='utf-8', engine='python')
df.drop(['Age'], axis=1, inplace=True)

x_train, x_test, y_train, y_test = train_test_split(df.drop(['Diabetes'], axis=1), df[['Diabetes']], test_size=0.2, shuffle=True, stratify=df[['Diabetes']], random_state=33)
df = pd.concat([x_train, y_train], axis=1).reset_index(drop=True)

df.head()
```

```python
from tensorflow.keras import utils
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
pca = PCA(n_components=13)

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)

# y_train = utils.to_categorical(y_train)
# y_test = utils.to_categorical(y_test)

y_train = pd.DataFrame(y_train).astype(int).values
y_test = pd.DataFrame(y_test).astype(int).values

X = x_train.copy()
Y = y_train.copy()
```

##### 인공신경망 기반

```python
from tensorflow.keras import utils, models, layers, initializers, optimizers, losses, metrics

model = models.Sequential()

model.add(layers.Dense(
    units = 300,
    activation = None,
    input_dim = x_train.shape[1],
    kernel_initializer = initializers.he_uniform,
))
model.add(layers.BatchNormalization())
model.add(layers.Activation('elu'))
model.add(layers.Dropout(0.2))


model.add(layers.Dense(
    units = 200,
    activation = None,
    kernel_initializer = initializers.he_uniform,
))
model.add(layers.BatchNormalization())
model.add(layers.Activation('elu'))
model.add(layers.Dropout(0.2))


model.add(layers.Dense(
    units = 100,
    activation = None,
    kernel_initializer = initializers.he_uniform,
))
model.add(layers.BatchNormalization())
model.add(layers.Activation('elu'))
model.add(layers.Dropout(0.2))


model.add(layers.Dense(
    units = 50,
    activation = None,
    kernel_initializer = initializers.he_uniform,
))
model.add(layers.BatchNormalization())
model.add(layers.Activation('elu'))
model.add(layers.Dropout(0.2))


model.add(layers.Dense(units = y_train.shape[1], activation='sigmoid'))


model.compile(
    optimizer = optimizers.Adam(),
    loss = losses.binary_crossentropy,
    metrics = [metrics.binary_accuracy, metrics.Recall(name='recall'), metrics.Precision(name='precision')]
)

history = model.fit(x_train, y_train, batch_size = 300, epochs=20, validation_split = 0.2)

y_pred = model.evaluate(x_test, y_test, batch_size = 30)
```

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import KFold
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

import matplotlib.pyplot as plt

validation = []
kfold = KFold(n_splits=5)

for train_idx, test_idx in kfold.split(x_train):
    x_train, x_val = X[train_idx], X[test_idx]
    y_train, y_val = Y[train_idx], Y[test_idx]

    model = GradientBoostingClassifier(
        # verbose = 1,
        random_state=33
    )

    y_pred = model.fit(x_train, y_train).predict(x_val)
    
    validation.append(accuracy_score(y_val, y_pred))
    validation.append(recall_score(y_val, y_pred))
    validation.append(precision_score(y_val, y_pred))
    validation.append(f1_score(y_val, y_pred))

    print()
    conf = ConfusionMatrixDisplay(confusion_matrix(y_val, y_pred))
    conf.plot()
    plt.show()

    print()
    print(f'accuracy score : {validation[0]}')
    print(f'recall score : {validation[1]}')
    print(f'precision score : {validation[2]}')
    print(f'f1 score : {validation[3]}')
    print()
    
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test)

conf = ConfusionMatrixDisplay(confusion_matrix=(confusion_matrix(y_test, y_pred)))
conf.plot()
plt.show()

print(f'accuracy_score : {accuracy_score(y_test, y_pred)}')
print(f'recall_score : {recall_score(y_test, y_pred)}')
print(f'precision_score : {precision_score(y_test, y_pred)}')
print(f'f1 score : {f1_score(y_test, y_pred)}')
print()
print(classification_report(y_test, y_pred))
```
