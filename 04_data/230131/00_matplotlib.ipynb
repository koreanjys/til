# python 의 graph 모듈 matplotlib

```python
import matplotlib.pyplot as plt


# 그림이 뜨는 형식을 지정
## %를 붙이는 매직명령어 (jupyter 에서만 사용)

# 라인 안에서 그리게 함
%matplotlib inline 

# 새 창에서 뜨게 함
# %matplotlib qt
```

```python
plt.plot([2, 3, 5, 10])  # tuple, ndarray, list 다 가능
plt.show()
```

```python
# x값, y값 넣어주기
# X값을 안 넣어주면 default 값 0부터 시작 0, 1, 2, 3~~
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.show()
```

### 스타일 지정

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], 'ro')  # ro 빨간색 점
plt.axis([0, 7, 0, 20])  # 범위 지정 [x_시작, x_끝, y_시작, y_끝]
plt.show()
```

```python
# 여러 개의 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np
data = np.arange(0,5, 0.2)
```

```python
# 빨간 대쉬, 녹색 사각형, 파란 삼각형
plt.plot(data, data, 'r--', data, data**2, 'gs', data,data**3, 'b^')
plt.show()
```
