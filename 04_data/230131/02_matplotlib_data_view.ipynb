### 여러 그래프

```python
# y1 = 2x^2
# y2 = 5x + 30
# y3 = 4x^2 + 10

import numpy as np

x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
```

```python
### 여러 개의 함수 그리기

import matplotlib.pyplot as plt

# 1번째

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
# 2번째

plt.plot(x,y1, x, y2, x, y3)
plt.show()
```

```python
### 새로운 그래프 창을 생성함

plt.plot(x,y1) # 처음 그리 함수를 수행하면 그래프 창이 자동으로 생성됨
 
plt.figure() # 새로운 그래프 창을 생성함
plt.plot(x,y2) # 새롭게 생성된 그래프 창에 그래프를 그림

plt.show()
```

```python
### 두 개 데이터 쌍 (x,y1), (x,y2) 를 지정된 번호의 그래프 창에 그래프로 그리기
### y1 = x^2 - 2, y2 = 20 cos(x)^2

import numpy as np

# 데이터 생성
x = np.arange(-5,5,0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2

plt.figure(1) # 1번 그래프 창을 생성한다
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그린다.

plt.figure(2) # 2번 그래프 창을 생성한다
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그린다.

plt.figure(1) # 이미 생성된 1번 그래프 창을 생성한다
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그린다.

plt.figure(2) # 이미 생성된 2번 그래프 창을 생성한다
plt.clf() # 2번 그래프 창에 그려진 모든 그래프를 지운다.
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그린다.

plt.show()
```

### subplot()을 활용하자

- 하나의 그래프 창을 하위 그래프 영역으로 나눈 후에 각 영역에 그래프를 그리는 방법

plt.subplot(m,n,p)  ### m * n 행렬로 이루어진 p번째 그래프

```python
import numpy as np

# 데이터 생성
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2 # sin 함수
y4 = 10*np.exp(-x) + 1 

# 2 X 2 행렬로 이뤄진 하위 그래프에서 p에 따라 위치를 지정
plt.subplot(2,2,1) # p는 1
plt.plot(x,y1)

plt.subplot(2,2,2) # p는 2
plt.plot(x,y2)

plt.subplot(2,2,3) # p는 3
plt.plot(x,y3)

plt.subplot(2,2,4) # p는 4
plt.plot(x,y4)

plt.show()
```

### 한글 깨짐 폰트 설정

```python
import matplotlib

matplotlib.rcParams['font.family'] ='Malgun Gothic' # 윈도우즈의 '맑은 고딕'으로 설정
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import numpy as np
x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
plt.plot(x,y1, '>--r', x,y2, 's-g', x,y3, 'd:b', x,y4, '-.Xc')
plt.legend(['데이터1','데이터2','데이터3','데이터4'], loc='best')
plt.xlabel('X 축')
plt.ylabel('Y 축')
plt.title('그래프 제목')
plt.grid(True)
```
